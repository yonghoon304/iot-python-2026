````python
# 이미지로드
img = cv2.imread('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_104204228.jpg')

cv2.imshow('origin',img) # origin 윈도우 제목으로 이미지 오픈
cv2.waitKey(0)  # 키보드 입력 대기, q키를 누르면 종료
cv2.destroyAllWindows() # 종료 후 열린 모든 윈도우창 닫음
````

````python
# 동영상 로드
cap = cv2.VideoCapture('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_112839302.mp4')

while True : # 무한 루프
    ret,frame = cap.read()  # fps의 이미지 하나씩 읽어옴
    if not ret : break  # 더 이상 읽을 이미지가 없으면 종료

    cv2.imshow('video',frame)   # 이미지를 계속 변경 
    if cv2.waitKey(1) ==ord('q'):break # 키보드 q입력하면 종료
cap.release() # 메모리 해제
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 변경
# 로드한 이미지는 매트릭스 객체에 할당
cat = cv2.imread('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_104204228.jpg',cv2.IMREAD_COLOR_BGR)

# openCV 함수로 변경
gray = cv2.cvtColor(cat,cv2.COLOR_BGR2GRAY) # 컬러를 회색톤으로 변경

cv2.imshow('origin',cat)
cv2.imshow('grey',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

````

````python
# 채널 분리
cat = cv2.imread('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_104204228.jpg',cv2.IMREAD_COLOR_BGR)

(b,g,r)= cv2.split(cat)

# 채널(B,G,R)로 단일 채널로 분리
# np.zeros() : 배열사이즈를 직접 지정해서 0으로 배열 초기화
# np.zeros_like(배열) : 지정된 배열사이즈를 0으로 초기화
blue = cv2.merge([b,np.zeros_like(g),np.zeros_like(r)])
green = cv2.merge([np.zeros_like(b),g,np.zeros_like(r)])
red = cv2.merge([np.zeros_like(b),np.zeros_like(g),r])


cv2.imshow('origin',cat)
cv2.imshow('blue',blue)
cv2.imshow('green',green)
cv2.imshow('red',red)

cv2.waitKey(0)
cv2.destroyAllWindows()

````

#### OpenCV 이미지 처리 
- 주요기능 - 포토샵 등 이미지편집기 기능과 동일
    - 이미지 읽기,저장,출력
    - 이미지 합성,이미지 자르기
    - 색상변환
    - 크기변경
    - 회전,이동
    - 블러
    - 이진화
    - 선명도 조절
    - 노이즈 제거
    - 대비,밝기 조절
    - 엣지 검출
    - 윤곽선 찾기
    - 퍼시팩티브 변환(원근감 -> 평탄화)


````python
# 이미지 로드
cat = cv2.imread('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_104204228.jpg',cv2.IMREAD_COLOR_BGR)

# 원그리기
# OpenCV 이미지 자체는 Mat라서 행,열 순으로 처리
# 원 포인트는 (x,y)로 처리
# thickess = -1은 내부 채우기
cv2.circle(cat,(416,210),100,(0,255,0),thickness=2,lineType=cv2.LINE_AA)


cv2.imshow('origin',cat)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

### ROI
- Region of Interest : 관심영역
    - 2차원리스트(매트릭스,행렬)
    - 영역추출법 - Mat[y시작점:y시작점 + 높이, x시작점: x시작점 + 너비]

````python
# 이미지 로드
cat = cv2.imread('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_104204228.jpg',cv2.IMREAD_COLOR_BGR)

# ROI
roi = cat[81:81+214,295:535]

cv2.imshow('origin',cat)
cv2.imshow('roi',roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드
cat = cv2.imread('C:\SourceBank\iot-python-2026\day05\KakaoTalk_20260401_104204228.jpg',cv2.IMREAD_COLOR_BGR)

# 엣지(이미지 경계) 검출
gray = cv2.cvtColor(cat,cv2.COLOR_BGR2GRAY)

# canny 엣지
# threshold 임계값, 1은 약한 엣지, 2는 강한 엣지
canny = cv2.Canny(gray,threshold1=100,threshold2=200)

cv2.imshow('edge',gray)
cv2.imshow('canny',canny)

cv2.waitKey(0)
cv2.destroyAllWindows()
````
