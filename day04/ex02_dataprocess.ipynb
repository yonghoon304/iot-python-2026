#### 기본 차트출력

````python
# matplotlib 라이브러리 로드
import matplotlib.pyplot as plt
````

````python
x = [1,2,3,4,5]
y=[10,20,15,30,25]

plt.plot(x,y)
plt.show()
````

#### 제목,레이블 표시
- 기본적으로 한글 출력 안됨

````python
plt.plot(x,y,marker='o')    # 마커표시
plt.title("점수 추이")
plt.xlabel("시간")
plt.ylabel("score")
plt.show()
````

- 한글 설정

````python
# 리소스인수,폰트관리자,리소스 포함
from matplotlib import rcParams,font_manager,rc

font_path =r'C:\Windows\Fonts\nanumgothiccoding.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font)
rcParams['axes.unicode_minus'] = False # 마이너스 표시 깨짐방짐
````

#### 라인,색상,포인트
- x,y 좌표 뒤에 문자열로 작성
    - 색상 : b,g,r,c,m,y,k,w
    - 마커 
    - 라인스타일 : -(line),--(dashed line),-.(dashed dot line),:(dotted line)
     

````python
plt.plot(x,y,'go-.')    
plt.title("점수 추이")
plt.xlabel("시간")
plt.ylabel("score")
plt.show()
````

#### 스타일지정
    - seaborn은 차트 생성전에 스타일 지정

#### 차트종류
- plot 라인차트
- bar 막대그래프
- barh 가로막대그래프
- hist 히스토그램  분포표
- scatter 산점도
- pie 파이차트 비율표시

#### 차트 팁 
- grid(True) : 맷플롯립만 사용하는 그리드 표시, seaborn에서는 영향 없음
- figure(figsize) 그래프 크기 설정

````python
sns.set_style("whitegrid")  # 단순 스타일지정
sns.set_theme(font=font,rc={'axes.unicode_minus':False}) # 한글깨짐
````

````python
# 막대 그래프
plt.bar(x,y)
plt.title("점수 추이")
````

````python
import numpy as np
scores = list(np.random.randint(50,100,12))
list(scores)
````

````python
plt.figure(figsize=(1,1))
plt.hist(scores,bins=5)
plt.title("학생 점수 분포표")
plt.xlabel("점수대")
plt.ylabel("학생 수")
plt.show()
````

````python
# 산점도
study_time=[1,2,3,4,5,6,7,8]
score=[50,55,65,72,78,81,93,98]

plt.figure(figsize=(10,5))
plt.scatter(study_time,score,marker='^',color='g')
plt.title("학습시간 대비 점수 상관관계")
plt.xlabel("학습시간")
plt.ylabel("점수")
plt.show()
````

#### 3D 그래프

````python
# 첫번째 쉘에 라이브러리 모두 정리하는게 좋음
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import numpy as np

# 리소스인수,폰트관리자,리소스 포함
from matplotlib import rcParams,font_manager,rc

font_path =r'C:\Windows\Fonts\nanumgothiccoding.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font)
rcParams['axes.unicode_minus'] = False # 마이너스 표시 깨짐방짐
````

````python
x= [1,2,3,4]
y=[10,20,15,30]
z=[5,10,8,12]

fig = plt.figure()  # 새로운 차트 선언해서 변수할당
ax = fig.add_subplot(111,projection = '3d')

ax.scatter(x,y,z)
ax.set_xlabel("X축")
ax.set_ylabel("Y축")
ax.set_zlabel("Z축")

plt.title("세 축 데이터")
plt.show()
````

````python
# 3D Surface
x = np.linspace(-5,5,50)    # -5~5까지 50개 간격으로 쪼갬
y = np.linspace(-5,5,50)    # -5~5까지 50개 간격으로 쪼갬

x,y =np.meshgrid(x,y)
z=x**2+y**2

fig =plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(x,y,z)

plt.show()
````
