#### 위경도 특성
- 소수점 4자리 이상 작성해야 제대로 된 위치를 잡을 수 있음

````python
# folium 라이브러리 로드
import folium
````

````python
loc=[35.116336,129.090216]
m= folium.Map(location=loc,zoom_start=20)   # 지도 객체 생성, m 변수 할당

# 마커를 생성해서 지도변수에 추가
folium.Marker(location=loc,
              tooltip='부산',
              popup='용당7관',
              icon=folium.Icon(icon='home')).add_to(m)

m
````
