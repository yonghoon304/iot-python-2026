#### 기본 사용

````python
# workcloud 라이브러리 로드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
````

````python
text = '''Python is my favorite programing langyage. data analysis. python ai ai ai
       deep learning, marchine learning, contemorary, ai skill 
    '''
wc = WordCloud().generate(text)
````

````python
plt.imshow(wc)
plt.axis("off") # 차트 x,y축 눈금 레이블 의미하는 객체
plt.show()
````
