### 분석용 기초 이론

리스트 컴프리헨션
- 기존의 리스트 사용법을 좀더 간결하고 효율적으로 사용 위해서

````python
# 기존방식
result = []
for i in range(10):
    result.append(i)
````

````python
result
````

````python
# 기존방식 2
result.clear()

for i in range(2,40,2):
    result.append(i)
````

````python
result
````

````python
# 리스트 컴프리헨션 방식 1 
result2 = [i for i in range(20)]
result2
````

````python
# 리스트 컴프리헨션 방식 2
result2 = [i for i in range(2,40,2)]
result2
````

````python
result2 = [i*2 for i in range(1,20)]
result2
````

````python
coord = [(x,y) for x in range(3) for  y in range (3)]
coord
````

````python
result2 = [ i for i in range(20) if i%2 ==0]
result2
````
