#### FPS 표시
- 대부분 30fps 사용함

````python
import time

# ../이후에 ctrl space
cap = cv2.VideoCapture(0) # 첫 번째 웹캠(0)

prev_time = 0   # 이전 프레임시간


while True:
    ret,img =cap.read() # 웹캠 화면은 하나씩 로드,ret(리턴결과),img(실제 이미지)
    if not ret : break

    curr_time = time.time() # 프레임 출력시간
    # fps계산
    fps = 1/(curr_time-prev_time) if prev_time != 0 else 0
    prev_time=curr_time

    # 출력할 텍스트
    fps_text = f'FPS:{int(fps)}'
    # img 위에 텍스트
    cv2.putText(img=img,
                text=fps_text,
                org=(10,30),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.9,
                color=(0,255,0),
                thickness=2)
    cv2.imshow('webcam',img)
    if cv2.waitKey(1)==ord('q') : break 
    

cap.release()   # 메모리 종료
cv2.destroyAllWindows()
````

#### 스냅샷 저장

````python
# ../이후에 ctrl space
import cv2
import numpy as np
import time

cap = cv2.VideoCapture(0) 

prev_time = 0 

while True:
    ret,img =cap.read() 
    if not ret : break

    curr_time = time.time() 
    fps = 1/(curr_time-prev_time) if prev_time != 0 else 0
    prev_time=curr_time


    fps_text = f'FPS:{int(fps)}'
   
    cv2.putText(img=img,
                text=fps_text,
                org=(10,30),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.9,
                color=(0,255,0),
                thickness=2)
    
    cv2.imshow('webcam',img)

    key=cv2.waitKey(1)
    if key==ord('q') :
         break 
    elif key==ord('s'): # 스냅샷 저장
         filename = time.strftime('snapshot_%y%m%d%H%M%S.jpg')
         cv2.imwrite(filename,img)
         print(f"{filename} 스냅샷 저장")
        

cap.release()  
cv2.destroyAllWindows()
````

#### 해상도 조정

#### 얼굴 검출
- OpenCV 내장된 기능
    - haarcascade_frontalface_default.xml
    - haarcascade_eye_tree_eyeglasses.xml
    - haarcascade_eye.xml
    - haarcascade_smile.xml
    - haarcascade_frontalcatface_extended.xml

````python
# Haar Cascade 파일 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')   # 얼굴데이터 따로
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_smile.xml')    # 미소

cap = cv2.VideoCapture(0) 

if not cap.isOpened(): 
     raise RuntimeError('웹캠 오픈 실패!')

# 적용된 해상값 확인
real_w=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0 
show_info = False # 정보 표시 여부

while True:
    ret,img =cap.read() 
    if not ret : break

    # FPS    
    curr_time = time.time() 
    fps = 1/(curr_time-prev_time) if prev_time != 0 else 0
    prev_time=curr_time

    if show_info :
        fps_text = f'FPS:{int(fps)}'
        res_text = f'{int(real_w)}x{int(real_h)}'

               
        cv2.putText(img=img,
                text=fps_text,
                org=(10,30),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.9,
                color=(0,255,0),
                thickness=2)
        cv2.putText(img=img,
                text=res_text,
                org=(10,60),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.9,
                color=(0,255,0),
                thickness=2)
    
    # image -> 검출할 이미지
    # scaleFactor : 이미지를 몇 배씩 줄여가며 탐색할지 지정하는 값, 1.1은 10%(속도저하,정확도상승)
    # 얼굴 1.1~1.3 적합, 미소는 1.5~18 적합
    # minNeighors : 후보영역을 진짜 객체로 인정하려면 주변에서 몇 번이나 겹쳐 검출되어야하는지 횟수
    faces = face_cascade.detectMultiScale(img,1.1,5) 
    
          
    # 검출된 물체를 표시
    for (x,y,w,h) in faces:    # x,y는 좌측상단좌표, w,h 너비와 높이
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),2)

        # 얼굴 영역만 잘라서 ROI(Region of Interect)
        face_roi = img[y:y+h,x:x+w]

        # 얼굴내부에서 미소 검출
        smiles = smile_cascade.detectMultiScale(face_roi,1.7,20)

        for (sx,sy,sw,sh) in smiles:
            cv2.rectangle(face_roi,(sx,sy),(sx+sw,sy+sh),(0,255,0),2)

    cv2.imshow('webcam',img)

    key=cv2.waitKey(1)
    if key==ord('q') :
        break 
    elif key==ord('s'): 
        filename = time.strftime('snapshot_%y%m%d%H%M%S.jpg')
        cv2.imwrite(filename,img)
        print(f"{filename} 스냅샷 저장")
    elif key==9 : # tab 키
        show_info = not show_info # true는 false로 false는 true로
        print(f'표시여부 {show_info    }')

cap.release()  
cv2.destroyAllWindows()
````

#### 얼굴 모자이크

````python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')   # 얼굴데이터 따로

cap = cv2.VideoCapture(0) 

if not cap.isOpened(): 
     raise RuntimeError('웹캠 오픈 실패!')

# 적용된 해상값 확인
real_w=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0 
show_info = False # 정보 표시 여부

while True:
    ret,img =cap.read() 
    if not ret : break

    # FPS    
    curr_time = time.time() 
    fps = 1/(curr_time-prev_time) if prev_time != 0 else 0
    prev_time=curr_time

    if show_info :
        fps_text = f'FPS:{int(fps)}'
        res_text = f'{int(real_w)}x{int(real_h)}'

               
        cv2.putText(img=img,
                text=fps_text,
                org=(10,30),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.9,
                color=(0,255,0),
                thickness=2)
        cv2.putText(img=img,
                text=res_text,
                org=(10,60),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.9,
                color=(0,255,0),
                thickness=2)
    

    faces = face_cascade.detectMultiScale(img,1.1,5) 
    
          
    # 검출된 물체를 표시
    for (x,y,w,h) in faces:    # x,y는 좌측상단좌표, w,h 너비와 높이
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),2)

        # 얼굴 영역만 잘라서 ROI(Region of Interect)
        face_roi = img[y:y+h,x:x+w]

        # 모자이크 처리
        small = cv2.resize(face_roi,(w//15,h//15),interpolation=cv2.INTER_LINEAR)   # 작게 줄이고
        mosaic = cv2.resize(small,(w,h),interpolation=cv2.INTER_NEAREST)    # 픽셀뭉개기

        img[y:y+h,x:x+w]=mosaic # ROI로 교체

    cv2.imshow('webcam',img)

    key=cv2.waitKey(1)
    if key==ord('q') :
        break 
    elif key==ord('s'): 
        filename = time.strftime('snapshot_%y%m%d%H%M%S.jpg')
        cv2.imwrite(filename,img)
        print(f"{filename} 스냅샷 저장")
    elif key==9 : 
        show_info = not show_info 
        print(f'표시여부 {show_info    }')

cap.release()  
cv2.destroyAllWindows()
````

````python
import cv2
import webbrowser
import time # 시간 제어를 위해 추가

qr_detector = cv2.QRCodeDetector()

cap = cv2.VideoCapture(0)

last_data = ''  # 마지막으로 읽은 데이터
last_open_time = 0  # 마지막으로 사이트를 연 시간
cool_down = 3  # 재오픈까지 필요한 쿨타임 (초 단위)

frame_count = 0
det_interval = 10 # 검사 주기 (너무 길면 반응이 느리니 조절하세요)

while True:
    ret, img = cap.read()
    if not ret:
        break

    frame_count += 1
    data = ''
    points = None

    # QR 감지
    if frame_count % det_interval == 0:
        data, points, _ = qr_detector.detectAndDecode(img)

    if data: # QR 데이터가 있는 경우
        # 1. 외곽선 그리기
        if points is not None:
            points = points.astype(int)
            for i in range(len(points[0])):
                pt1 = tuple(points[0][i])
                pt2 = tuple(points[0][(i+1) % len(points[0])])
                cv2.line(img, pt1, pt2, (0, 255, 0), 2)
        
        cv2.putText(img, data, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        # 2. 웹사이트 오픈 로직 (쿨타임 및 중복 체크)
        current_time = time.time()
        if (data.startswith('http://') or data.startswith('https://')):
            # 데이터가 바뀌었거나, 같은 데이터라도 설정한 쿨타임이 지났다면 실행
            if data != last_data or (current_time - last_open_time) > cool_down:
                webbrowser.open(data)
                last_data = data
                last_open_time = current_time
                print(f"Opened: {data}")

    # 기존의 else 문(상태 초기화)을 삭제했습니다.
    # 그래야 QR이 화면에 계속 있어도 last_data가 유지되어 중복 실행을 막습니다.

    cv2.imshow('QR CODER', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
````
